```python
import pandas as pd
```

```python
# Load the data
data = pd.read_csv('../bcwd.csv')
data.head()
```

```python
# Normalize the data between 0 and 1 and column-wise
def normalize(col):
    return (col - col.min()) / (col.max() - col.min())

data = data.apply(normalize, axis=0)
```

```python
data.describe()
```

```python
# Split the data into training and testing
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(data.drop('Class', axis=1), data['Class'], test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier

def train_kn(xtrain, ytrain):
    md = KNeighborsClassifier(n_neighbors=3)
    md.fit(xtrain, ytrain)
    return md
```

```python
model = train_kn(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

```python
from sklearn.metrics import classification_report

print(classification_report(ytest, model.predict(xtest)))
```

```python
# Adding SMOTE to generate synthetic samples
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE()
xtrain_smote, ytrain_smote = smote.fit_resample(xtrain, ytrain)
```

```python
data0 = data[data['Class'] == 0]
xtrain, xtest, ytrain, ytest = train_test_split(data0.drop('Class', axis=1), data0['Class'], test_size=0.2, random_state=42)
train0data = pd.concat([xtrain, ytrain], axis=1)
```

```python
synthesizer.fit(train0data)
```

```python
synthetic_data = synthesizer.sample(num_rows=200)
```
